```python
from challenge_data import PATIENTS, to_pipeline_format
from processor import build_fhir_bundle, generate_clinical_snapshot

print(f"Loaded {len(PATIENTS)} patients: {[p['name'] for p in PATIENTS]}")
```

```python
# Convert all three patients to the flat format the pipeline expects
patients = [to_pipeline_format(p) for p in PATIENTS]

for p in patients:
    print(f"{p['subject_id']:20s}  {p['name']:26s}  steps/day: {p['mean_daily_steps']:>5,}  RHR: {p['resting_hr_bpm']} bpm")
```

```python
# Run the full pipeline for all three patients
results = []

for patient in patients:
    print(f"Processing {patient['name']}...")
    bundle   = build_fhir_bundle(patient)
    snapshot = generate_clinical_snapshot(patient, bundle)
    results.append({"patient": patient, "bundle": bundle, "snapshot": snapshot})
    print(f"  ✓ {len(bundle.entry)} FHIR observations generated")

print("\nAll patients processed.")
```

```python
def show_patient(r):
    p = r["patient"]
    print(f"{'='*60}")
    print(f"Patient : {p['name']}  ({p['subject_id']})")
    print(f"DOB     : {p['dob']}")
    print(f"Context : {p['clinical_context']}")
    print(f"{'─'*60}")
    print(f"Steps/day        : {p['mean_daily_steps']:,}")
    print(f"Active mins total : {p['total_active_minutes']}")
    print(f"Sedentary/day     : {p['mean_daily_sedentary_minutes']} min")
    print(f"Days on target    : {p['days_meeting_150min_activity_target']}/7")
    print(f"Mean HR           : {p['mean_heart_rate_bpm']} bpm")
    print(f"Resting HR        : {p['resting_hr_bpm']} bpm")
    print(f"{'─'*60}")
    print(r["snapshot"])

# Patient 1 — Mrs Margaret Whitfield
show_patient(results[0])
```
